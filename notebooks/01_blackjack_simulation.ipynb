## 🤖 AI Blackjack Strategy Implementation


```python
class HiLoCounter:
    """Hi-Lo Card Counting System"""
    
    def __init__(self, decks=6):
        self.running_count = 0
        self.true_count = 0
        self.decks_remaining = decks
        self.decks_initial = decks
        
        # Card values for Hi-Lo
        self.card_values = {
            '2': 1, '3': 1, '4': 1, '5': 1, '6': 1,
            '7': 0, '8': 0, '9': 0,
            '10': -1, 'J': -1, 'Q': -1, 'K': -1, 'A': -1
        }
    
    def update_count(self, cards):
        """Update running count with visible cards"""
        for card in cards:
            self.running_count += self.card_values.get(card, 0)
        
        # Update decks remaining
        cards_used = len(cards)
        self.decks_remaining = max(0.5, self.decks_initial - (cards_used / 52))
        
        # Calculate true count
        if self.decks_remaining > 0:
            self.true_count = self.running_count / self.decks_remaining
        else:
            self.true_count = 0
    
    def get_bet_multiplier(self):
        """Calculate bet multiplier based on true count"""
        if self.true_count <= 0:
            return 1
        elif self.true_count <= 2:
            return 2
        elif self.true_count <= 4:
            return 4
        elif self.true_count <= 6:
            return 6
        else:
            return 8

# Test the counter
counter = HiLoCounter(decks=6)
print('✅ Hi-Lo Counter initialized')
print(f'Running count: {counter.running_count}')
print(f'True count: {counter.true_count:.2f}')
```
